```python
import pandas as pd
import numpy as np
from textblob import TextBlob
import requests

# Load financial data
def load_financial_data(file_path):
    try:
        data = pd.read_csv(file_path)
        data.rename(columns={
            'Total Revenue (in millions)': 'Total Revenue',
            'Net Income (in millions)': 'Net Income',
            'Cash Flow from Operating Activities (in millions)': 'Cash Flow'
        }, inplace=True)
        data.fillna(0, inplace=True)
        return data
    except Exception as e:
        print(f"Error loading data: {e}")
        return None

# Fetch real-time news using NewsAPI
def fetch_real_time_news(company):
    api_key = "e4a0ad7c65f84f0196c4a11f2312007f"  # Your NewsAPI key
    url = f"https://newsapi.org/v2/everything?q={company}&sortBy=publishedAt&apiKey={api_key}"
    try:
        response = requests.get(url)
        if response.status_code != 200:
            print(f"Error fetching news for {company}: {response.status_code}")
            return []
        articles = response.json().get('articles', [])
        return [{"headline": article['title'], "url": article['url'], "polarity": analyze_sentiment_score(article['title'])} for article in articles[:5]]
    except Exception as e:
        print(f"Failed to fetch news for {company}: {e}")
        return []

# Sentiment analysis
def analyze_sentiment_score(text):
    try:
        analysis = TextBlob(text)
        return analysis.sentiment.polarity  # Sentiment polarity: [-1, 1]
    except Exception as e:
        return 0

# Financial performance analysis
def financial_performance_analysis(company_data):
    summary = []
    for _, row in company_data.iterrows():
        summary.append(f"- Year {row['Year']}: Revenue ${row['Total Revenue']}M, Net Income ${row['Net Income']}M, Cash Flow ${row['Cash Flow']}M")
    return "\n".join(summary)

# Risk scoring logic
def assess_risk(news_data, financial_features):
    avg_sentiment = np.mean([article['polarity'] for article in news_data]) if news_data else 0
    negative_articles = sum(1 for article in news_data if article['polarity'] < 0)
    revenue_growth = financial_features.get('revenue_growth', 0)
    net_income_growth = financial_features.get('net_income_growth', 0)

    risk_score = (
        -avg_sentiment * 0.5 +
        -revenue_growth * 0.3 +
        -net_income_growth * 0.2
    )

    if risk_score > 0.5:
        risk_level = "High"
    elif 0.2 < risk_score <= 0.5:
        risk_level = "Moderate"
    else:
        risk_level = "Low"

    return {
        "risk_level": risk_level,
        "risk_score": risk_score,
        "avg_sentiment": avg_sentiment,
        "negative_articles": negative_articles
    }

# Chatbot logic
def financial_chatbot(query, financial_data, context):
    query = query.lower()
    response = ""

    # Help command
    if "help" in query:
        return ("Hi! Here's how I can assist you:\n"
                "- Ask about 'financial performance' for a company.\n"
                "- Ask about 'risk level' or 'risk assessment'.\n"
                "- Combine queries like 'Show Tesla's financial performance and risks'.")

    # Parse query for company
    for company in financial_data['Company'].unique():
        if company.lower() in query:
            context['company'] = company
            break

    if not context.get('company'):
        return "Please specify the company you're inquiring about."

    company = context['company']
    company_data = financial_data[financial_data['Company'] == company]

    if "financial performance" in query:
        response += f"Financial Performance Analysis for {company}:\n"
        response += financial_performance_analysis(company_data)
        response += "\n"

    if "risk" in query or "risk assessment" in query:
        news_data = fetch_real_time_news(company)
        financial_features = {
            'revenue_growth': company_data['Total Revenue'].pct_change().iloc[-1] if len(company_data) > 1 else 0,
            'net_income_growth': company_data['Net Income'].pct_change().iloc[-1] if len(company_data) > 1 else 0,
            'avg_cash_flow': company_data['Cash Flow'].mean()
        }
        risk_result = assess_risk(news_data, financial_features)
        response += f"Risk Assessment for {company}:\n"
        response += f"- Risk Level: {risk_result['risk_level']}\n"
        response += f"- Risk Score: {risk_result['risk_score']:.2f}\n"
        response += f"- Average Sentiment: {risk_result['avg_sentiment']:.2f}\n"
        response += f"- Negative Articles Count: {risk_result['negative_articles']}\n"
        response += "\n### Top 5 News Articles:\n"
        for i, article in enumerate(news_data, 1):
            response += f"{i}. {article['headline']} (Sentiment: {article['polarity']:.2f})\n"
            response += f"   URL: {article['url']}\n"

    if not response:
        response = "I'm sorry, I couldn't understand your query. Please try asking about financial performance, risks, benchmarking, or investment support."

    return response

# Main function
if __name__ == "__main__":
    file_path = "/content/drive/MyDrive/Project Work/Combined_Financial.csv"
    financial_data = load_financial_data(file_path)

    if financial_data is not None:
        print("Hi! Welcome to the Financial Chatbot!")
        print("Type 'exit' to end the conversation or 'help' for assistance.")

        # Context to retain company
        context = {'company': None}

        while True:
            user_query = input("\nYour question: ")
            if user_query.lower() == "exit":
                print("Goodbye! Have a great day!")
                break
            print(financial_chatbot(user_query, financial_data, context))
    else:
        print("Failed to load financial data. Please check the file path or format.")
```
